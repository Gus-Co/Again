```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

years = list(map(int, range(1980,2014))) # Obtener todos los años necesarios del dataframe en las columnas.
df_can = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)


index_columns = ["OdName", "AreaName", "RegName", "DevName"] # Columnas necesarias como index de la pivot

rename_columns = {"OdName": "Country", "AreaName": "Continent", "RegName": "Region"} # Columnas a cambiar de nombre.

df_canada = df_can.pivot_table(index=index_columns, values=years)

df_canada.reset_index(inplace=True)

df_canada.rename(columns=rename_columns, inplace=True)

df_canada["Total"] = df_canada[years].sum(axis=1)

df_canada.set_index("Country", inplace=True)

df_canada.index.name = None

df_canada.sort_values(['Total'], ascending = False, axis = 0, inplace = True)

df_top5 = df_canada.head()
df_top5 = df_top5[years].transpose


# df_canada.head(3)

# # count, bin_edges = np.histogram(df_canada[2013])
# # df_canada[2013].plot(kind='hist', xticks = bin_edges)

# df_canada[2013].plot(kind='hist')

# plt.title('Histogram of Inunigration from 195 countries in 2013')
# plt.ylabel('Countries')
# plt.xlabel('Imigrants')


# # df_top5.plot(kind='area')

# plt.title('Im good')
# plt.ylabel('Number of imigrants')
# plt.xlabel('Years')
df_iceland = df_canada.loc['Iceland' , years]

df_iceland.plot(kind='bar')
plt.title( 'Icelandic irnmigrants to Canada from 1980 to 2013' )
plt.xlabel( 'Year' )
plt.ylabel( 'Number of immigrants' )
plt.show( )

plt.show()
```
